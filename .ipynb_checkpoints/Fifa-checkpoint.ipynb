### Clustering de Jugadores de Fifa

vamos a sumergirnos en el mundo del FIFA 18 y analizar un dataset de jugadores. ¿Se acuerdan de ese juego de consola? Es donde los jugadores de las ligas más importantes del mundo están súper detallados con un montón de características, como qué tan ágiles son, qué tan buenos en los tiros libres, su edad, de dónde son, cuánto ganan, y hasta su posición en el campo.


Algunas de las características que trae este dataset son:

Name: El nombre del jugador, obvio.
Age: Su edad.
Nationality: De qué país es.
Club: En qué equipo juega.
Value: Precio de mercado del jugador
Wage: Salario de los jugadores

Acceleration: Qué tan rápido arranca.
Agility: Su agilidad, qué tan fácil se mueve.
Balance: Su equilibrio.
Ball control: Qué tan bien controla el balón.
Free kick accuracy: Qué tan precisos son en los tiros libres.
Y claro, las posiciones en el campo también están representadas, por ejemplo:

CAM (Centrocampista Ofensivo Central)
CB (Defensa Central)
CDM (Centrocampista Defensivo Central)
CF (Delantero Central)
CM (Centrocampista Central)

Nuestro gran objetivo aquí es ver si podemos encontrar alguna estructura o "grupos" de jugadores en estos datos. La idea es aplicar técnicas de clustering y, una vez que tengamos esos grupos, tratar de entender qué es lo que los une, es decir, ¡cuál es el criterio que mejor los agrupa! El dataset es extraido de [acá](https://www.kaggle.com/thec03u5/fifa-18-demo-player-dataset)

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../Clustering y Jugadores de Fifa/Fifa_players.csv', low_memory=False)
```

```python
data
```

###  Limpieza

Eliminaremos los jugadores con valor cero y la convertiremos la columna de valor a numérica eliminando los euros, trasnformando los M en millones y K en mil. 



```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
     display(data.head(3))
     display(data.dtypes)
```

```python
# Elimino algunas columnas que veo no necesarias 
data=data.drop(['Unnamed: 0','Photo','Flag','Club Logo'],axis=1)
```

```python
# Cuento los valores de los campos Value y Wage
print(data['Value'].apply(lambda x: x[0]).value_counts())
print(data['Wage'].apply(lambda x: x[0]).value_counts())
```

```python
# Reviso la terminación del string en el campo valores
data['Value'].apply(lambda x: x[-1]).value_counts()
```

```python
data['Wage'].apply(lambda x: x[-1]).value_counts()
```

```python
# Limpio el campo 'Value'
data['Value'] = data['Value'].str.replace('€', '', regex=False)

# Máscaras para M (millones) y K (miles)
M_mask = data['Value'].apply(lambda x: x[-1] == 'M')
K_mask = data['Value'].apply(lambda x: x[-1] == 'K')

# Convierto los valores con sufijo
data.loc[M_mask, 'Value'] = data.loc[M_mask, 'Value'].apply(lambda x: float(x[:-1]) * 1e6)
data.loc[K_mask, 'Value'] = data.loc[K_mask, 'Value'].apply(lambda x: float(x[:-1]) * 1e3)

# El resto (sin letra al final) lo paso a float
rest_mask = ~(M_mask | K_mask)
data.loc[rest_mask, 'Value'] = data.loc[rest_mask, 'Value'].astype(float)
```

```python
# Limpio el campo 'Wage'
data['Wage'] = data['Wage'].str.replace('€', '', regex=False)

# Máscara para valores con 'K' al final (miles)
K_mask = data['Wage'].apply(lambda x: x[-1] == 'K')

# Convierto los que tienen 'K'
data.loc[K_mask, 'Wage'] = data.loc[K_mask, 'Wage'].apply(lambda x: float(x[:-1]) * 1e3)

# El resto los paso a float
rest_mask = ~K_mask
data.loc[rest_mask, 'Wage'] = data.loc[rest_mask, 'Wage'].astype(float)
```

```python
# Eliminamos los jugadores con valor = 0

data=data.drop(data[data['Value']==0].index,axis=0);
```

```python
# Preparamos el dataset para clustering

data_clustering=data[['Acceleration', 'Aggression', 'Agility', 'Balance',
       'Ball control', 'Composure', 'Crossing', 'Curve', 'Dribbling',
       'Finishing', 'Free kick accuracy', 'GK diving', 'GK handling',
       'GK kicking', 'GK positioning', 'GK reflexes', 'Heading accuracy',
       'Interceptions', 'Jumping', 'Long passing', 'Long shots', 'Marking',
       'Penalties', 'Positioning', 'Reactions', 'Short passing', 'Shot power',
       'Sliding tackle', 'Sprint speed', 'Stamina', 'Standing tackle',
       'Strength', 'Vision', 'Volleys']]
```

```python
# Convertimos las columnas a variables numericas y eliminamos las filas que no necesitamos

object_columns=[col for col in data_clustering.columns if data[col].dtype=='object']

mask=data_clustering[object_columns].applymap(lambda x:  not str(x).isnumeric())
data_clustering=data_clustering.drop(data_clustering[mask.sum(axis=1)>0].index,axis=0)
```

### Clustering



```python
scaler=StandardScaler()
X=scaler.fit_transform(data_clustering)

inertia=[]
sil=[]
ch_scores=[]
k_values=range(2,20,2)

for k in k_values:
    print(k)
    km=KMeans(n_clusters=k)
    km.fit(X)
    inertia.append(km.inertia_)    
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='#5e60ce', label='Inercia')  # violeta bonito
plt.vlines(4, ymin=min(inertia), ymax=max(inertia), colors='#f72585', linestyle='dashed', label='k óptimo = 4')  # rosado fuerte
plt.xlabel('Número de clusters (k)', fontsize=14, color='#3a0ca3')
plt.ylabel('Inercia', fontsize=14, color='#3a0ca3')
plt.title('Método del codo para determinar k óptimo', fontsize=16, color='#3f37c9')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

### Composición de los clusters


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual de seaborn
sns.set(style='whitegrid', palette='pastel', font_scale=1.1)

# Entrenar el modelo KMeans
k = 4
km = KMeans(n_clusters=k, random_state=0)
km.fit(X)

# Filtrar el dataset original
data_filtered = data.loc[data_clustering.index]

# Crear subplots
f, ax = plt.subplots(2, 2, figsize=(16, 10), gridspec_kw={'wspace': 0.4, 'hspace': 0.4})
ax = ax.ravel()

# Colores por cluster
colors = ['#5e60ce', '#4cc9f0', '#f72585', '#b5179e']

# Graficar los top 10 valores por cluster
for lbl in range(k):
    S = data_filtered.loc[km.labels_ == lbl, 'Preferred Positions'].value_counts(normalize=True)
    chart = sns.barplot(y=S.index[:10], x=S.values[:10], ax=ax[lbl], orient='h', color=colors[lbl])
    chart.set_title(f'Cluster {lbl}', fontsize=14, color=colors[lbl])
    chart.set_xlabel('Proporción', fontsize=12)
    chart.set_ylabel('Posición preferida', fontsize=12)
    chart.tick_params(axis='x', labelsize=10)
    chart.tick_params(axis='y', labelsize=10)

# Título general
f.suptitle('Distribución de posiciones preferidas por cluster', fontsize=18, color='#3a0ca3')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # espacio para el título
plt.show()
```

```python
grupos = [[0,12], [1,2], [3,9,14], [4,5,6,8], [7,11], [10,13]]

f, ax = plt.subplots(3, 2, figsize=(15, 10), gridspec_kw={'wspace': 0.5})
ax = ax.ravel()

for i, g in enumerate(grupos):
    for c in g:
        # Extrae valores numéricos positivos
        values = data_filtered.loc[km.labels_ == c, 'Value']
        values = pd.to_numeric(values, errors='coerce')  # Asegura tipo numérico
        values = values.dropna()  # Elimina NaNs
        values = values[values > 0]  # Evita log(0) o log negativo

        # Aplica logaritmo
        log_values = np.log(values)

        # Grafica
        sns.kdeplot(log_values, label=f'Cluster {c}', ax=ax[i])
        
    ax[i].legend()
    ax[i].set_title(f'Grupos: {g}', fontsize=12)
    ax[i].set_xlabel('Log(Value)')
    ax[i].set_ylabel('Densidad')

plt.suptitle('Distribución logarítmica del valor por grupos de clusters', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
